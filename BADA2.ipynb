```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
emp_df = pd.read_csv('/content/Emp_engagement.csv')
print(emp_df)
```

```python
emp_df.head()
```



---



```python
emp_df['Engagement'] = emp_df[['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9']].mean(axis=1)
emp_df['Skill Variety'] = emp_df[['S1', 'S2', 'S3', 'S4']].mean(axis=1)
emp_df['Autonomy'] = emp_df[['A1', 'A2', 'A3']].mean(axis=1)
emp_df['Task Significance'] = emp_df[['TS1', 'TS2', 'TS3']].mean(axis=1)
emp_df['Task Identity'] = emp_df[['TI1', 'TI2', 'TI3', 'TI4']].mean(axis=1)
emp_df['Feedback'] = emp_df[['F1', 'F2', 'F3']].mean(axis=1)
emp_df['Satisfaction'] = emp_df[['SAT1', 'SAT2', 'SAT3', 'SAT4', 'SAT5']].mean(axis=1)
print(emp_df)
```

```python
emp_df.head()
```



---



```python
emp_df = emp_df.drop(['Gender','Ageanization','Education', 'Tenure', 'Organization','Serial number', 'E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'S1', 'S2', 'S3', 'S4', 'A1', 'A2', 'A3', 'TS1', 'TS2', 'TS3', 'TI1', 'TI2', 'TI3', 'TI4', 'F1', 'F2', 'F3', 'SAT1', 'SAT2', 'SAT3', 'SAT4', 'SAT5'],axis=1)
emp = emp_df.dropna(axis=0,how="all")
print(emp)
```

```python
emp.head()
```



---



```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
emp = sc.fit_transform(emp)
```



---



```python
print("Is the dataframe 'emp' NaN?\n" +'\033[1m',np.any(np.isnan(emp)))
print ('\033[0m')
print("Are all the elements finite?\n"+ '\033[1m',np.all(np.isfinite(emp)))
```



---



```python
from sklearn.decomposition import PCA
pcs = PCA()
pcs.fit_transform(emp)
plt.plot(range(7), pcs.explained_variance_ratio_, color='#3f3769')
plt.plot(range(7),np.cumsum(pcs.explained_variance_ratio_),color='#c20404')
plt.title("No. of Components vs Cummulative Explained Variance ratio", color='#176e03')
```



---



# ***Conclusion:***

*We initially have a dataset with about 37 columns, which we reduce to 7 columns by first calculating the average of specific rows to form new columns such as average of A1, A2 and A3 gives us the column autonomy. Similarly it is perfomed for a total of 7 variables as such.* <br> <br>
*From the resulting graph, we notice that the maximum number of components upto which the data of 7 columns can be reduced to give an **explained variance ratio of 90%** will be **5 or 6**.*. <br><br>
*So, while doing any prediction using **Principle Component Analysis(PCA)**to get the number of components, reduction to 5 or 6 will give an accurate value.*
